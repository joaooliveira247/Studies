```python
from matplotlib import pyplot as plt


def simple_graf(x_label: list[int], y_label: list[float]) -> None:
    plt.plot(
        *[x_label, y_label],
        **{"color": "green", "marker": "o", "linestyle": "solid"}
    )
    plt.title("Nominal PIB")
    plt.ylabel("Bilhões de Dolares")
    plt.show()

def main() -> None:
    years: list[int] = [1950, 1960, 1970, 1980, 1990, 2000, 2010]

    gdp: list[float] = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

    simple_graf(years, gdp)


if __name__ == "__main__":
    main()
```
