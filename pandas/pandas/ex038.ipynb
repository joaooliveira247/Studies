## Gráficos de barras com as frequências das

```python
import pandas as pd

flags = pd.read_csv("./csv/flags.csv")

df_cores = pd.DataFrame()

for c in flags.columns:
    if c in ['red','green','blue','gold','white','black','orange']:
        df_cores[c] = flags[c].value_counts()

lst_cores = [
    ["red", "beige"],
    ["green", "beige"],
    ["blue", "beige"],
    ["gold", "beige"],
    ["whitesmoke", "beige"],
    ["black", "beige"],
    ["orange", "beige"],
]

df_cores.plot(
    kind="barh",
    subplots=True,
    figsize=(8, 25),
    color=lst_cores
)
```
